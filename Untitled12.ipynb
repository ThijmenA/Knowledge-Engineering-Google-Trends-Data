```python
import pandas as pd

file_path = "C:/data/rainfile.txt"  

df = pd.read_csv(file_path, comment='#', sep=',', header=None, names=['STN', 'YYYYMMDD', 'RH'])

df['DATE'] = pd.to_datetime(df['YYYYMMDD'], format='%Y%m%d')

df['RH'] = df['RH'].replace(-1, pd.NA)

df['RH'] = df['RH'] / 10.0
```

```python
df['YEAR_MONTH'] = df['DATE'].dt.to_period('M').astype(str)

monthly_avg = df.groupby('YEAR_MONTH')['RH'].mean().reset_index()

monthly_avg.to_csv('monthly_avg_precipitation.csv', index=False)

monthly_avg
```
