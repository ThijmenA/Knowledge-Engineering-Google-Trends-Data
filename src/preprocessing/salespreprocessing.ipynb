```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/sales_data_2.csv", sep=";")
data.head()
```

```python
data["SectorBranchesSIC2008"].unique()
```

```python
data["TurnoverIndicesOnlineSales_1"] = data["TurnoverIndicesOnlineSales_1"].str.strip().replace(".", pd.NA)
data
```

```python
sectors = ["900049 ", "900058 ", "382600 ", "B000625", "B000626", "B000627", "B000628"]
translated_sectors = ["Retail trade", "Multi-channel", "Retail sale via internet", "Retail sale of food and drugstore items", "Retail sale of consumer electronics", "Retail sale of clothes and fashion items", "Retail sale of other non-food"]
for sector, translated_sector in zip(sectors, translated_sectors):
    data.loc[data["SectorBranchesSIC2008"] == sector, "SectorBranchesSIC2008"] = translated_sector
data
```

```python
pivot_df = data.pivot(index='Periods', columns='SectorBranchesSIC2008', values='TurnoverIndicesOnlineSales_1')
pivot_df
```

```python
pivot_df.index = pivot_df.index.str.replace("MM", "-")
pivot_df
```

```python
pivot_df[~pivot_df["Retail sale of other non-food"].isnull()]
```

```python
pivot_df.to_csv("../data/sales_data_2_transformed.csv")
```
